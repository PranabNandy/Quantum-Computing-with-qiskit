# Different Quantum Gates in Quantum Circuits

```python
from qiskit import QuantumRegister, ClassicalRegister, Aer, execute, QuantumCircuit
from qiskit.tools.visualization import plot_histogram
```

```python
S_simulator = Aer.get_backend(name='statevector_simulator')
Q_simulator = Aer.get_backend(name='qasm_simulator')
```

```python
q = QuantumRegister(2)
c = ClassicalRegister(2)
```

# Identity Gate

```python
ckt = QuantumCircuit(q,c)

ckt.id(q[0])
ckt.measure(q, c)

ckt.initialize([0, 1], 0) 
ckt.initialize([1, 0], 1)

ckt.draw(output = "mpl")
```

```python
# Execute the circuit
ex = execute(ckt, Q_simulator, shots = 1024)
res = ex.result()
bars = res.get_counts(ckt)

# Plot the results
plot_histogram(bars, bar_labels=False)
```

# Identity Gate

```python
#initialize circuit
ckt = QuantumCircuit(q,c)

ckt.initialize([0, 1], 0) #first bit 1
ckt.initialize([1, 0], 1) #second bit 0

# circuit design
ckt.x(q[1])
ckt.measure(q, c)

#plot circuit
ckt.draw(output = "mpl")
```

```python
# Execute the circuit
ex = execute(ckt, Q_simulator, shots = 1024)
res = ex.result()
bars = res.get_counts(ckt)

# Plot the results
plot_histogram(bars, bar_labels=False)
```

# CX Gate

```python
#initialize circuit
ckt = QuantumCircuit(q,c)

ckt.initialize([0, 1], 0) #first bit 1
ckt.initialize([1, 0], 1) #second bit 0

# circuit design
ckt.cx(q[0], q[1])
ckt.measure(q, c)

#plot circuit
ckt.draw(output = "mpl")
```

```python
# Execute the circuit
ex = execute(ckt, Q_simulator, shots = 1024)
res = ex.result()
bars = res.get_counts(ckt)

# Plot the results
plot_histogram(bars, bar_labels=False)
```
